```python
import numpy as np
from sklearn.datasets import *
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import mglearn
import seaborn as sns
```

```python
cancer_data = load_breast_cancer()
cancer_data
```

```python
df_cancer = pd.DataFrame(
    np.c_[cancer_data['data'],cancer_data['target']],
          columns = np.append(cancer_data['feature_names'], ['target'])
          )
```

```python
cancer_data.feature_names
```

```python
len(df_cancer.columns)
```

```python
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area'])#, 'mean perimeter', 'mean smoothness'] )
```

```python
plt.figure(figsize=(20,10)) 
sns.heatmap(df_cancer.corr(), annot=True, cmap = 'mako') 
```

```python
X = df_cancer.drop('target', axis =1)
y = pd.DataFrame(cancer_data.target)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 8)
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

```python
y_pred = svc_model.predict(X_test)
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm,annot=True)
```

```python
#bad results
```

Normalization

```python
X_train = (X_train - X_train.min())/((X_train - X_train.min()).max())
```

```python
X_test = (X_test - X_test.min())/((X_test - X_test.min()).max())
```

```python
svc_model = SVC()
svc_model.fit(X_train, y_train)
y_pred = svc_model.predict(X_test)
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm,annot=True)
```
